```python
import zipfile
import os
import pandas as pd

# FIND MAXIMUM OF THE DATA SETS To be able to Normaliz the Whole Tumor Dimensions

maxList = []
minList = []

# Construct the correct path to the zip file
current_dir = os.getcwd()
zip_file_path = os.path.join(current_dir, '..', 'In vivo data', 'In vivo data.zip')

# Unzip the file
extracted_dir = os.path.join(current_dir, '..', 'In vivo data')
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extracted_dir)

# Load the data and filter it
file_mapping = {
    'Study1': 'Study_1_1.xlsx',
    'Study2': 'Study_2_1.xlsx',
    'Study3': 'Study_3_1.xlsx',
    'Study4': 'Study_4_1.xlsx',
    'Study5': 'Study_5_1.xlsx'
}

filtered_data_frames = {}

for studyName in studies:
    file_name = file_mapping.get(studyName)
    if file_name:
        file_path = os.path.join(extracted_dir, file_name)
        data, arms = utils.Read_Excel(file_path, ArmName='TRT01A', split=splits[studies.index(studyName)])
        filtered_Data = data.loc[data['TRLINKID'] == 'INV-T001']
        filtered_Data = filtered_Data.loc[filtered_Data['TRTESTCD'] == 'LDIAM']
        temp = list(filtered_Data['TRORRES'])
        temp = utils.Remove_String_From_Numeric_Vector(temp, valueToReplace=0)
        if temp:  # Check if temp is not empty
            maxList.append(max(temp))
            minList.append(min(temp))
        # Store the filtered data frame in the dictionary
        filtered_data_frames[f'df_{studyName.lower()}_filtered'] = filtered_Data
    else:
        print(f"File for {studyName} not found in the extracted files.")

# Display the names of the filtered data frames
print("Names of the filtered data frames:", list(filtered_data_frames.keys()))

# Access and display the filtered data frames
for df_name, df in filtered_data_frames.items():
    print(f"Filtered data frame: {df_name}")
    display(df.head())  # Display the first few rows of each filtered data frame
```
